- bs4 í™œìš©í•˜ì—¬ íŒŒì‹± í›„ í…ìŠ¤íŠ¸ ë°ì´í„° ì¶”ì¶œ

````python
from bs4 import BeautifulSoup as bs
with open("test.xml", 'r') as f:
    xml_doc = f.read()
soup = bs(xml_doc, 'xml')
for row in soup.find_all('people'):
    print(row.get('name'))
    print(row.age.text)
    print(row.tall.text)
    print(row.add.text)
print(soup.world.naver_api.client_id.text, soup.world.naver_api.client_secret.text)
````

- UTF-8ì´ ì•„ë‹Œ EUC-KRì´ë‚˜ cp949ì¼ê²½ìš° decode('cp949')ë¡œ ë³€ê²½
- urllibê³¼ bs4ì˜ find ë¬¸ í™œìš©í•œ ì›¹ í¬ë¡¤ë§ ë°©ì‹

````python
# urllib ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œ urlopen í•¨ìˆ˜ë¥¼ import
from urllib.request import urlopen

# BeautifulSoup ëª¨ë“ˆì—ì„œ BeautifulSoup í´ë˜ìŠ¤ë¥¼ bsë¼ëŠ” ì´ë¦„ìœ¼ë¡œ import
from bs4 import BeautifulSoup as bs

# ì›¹ í˜ì´ì§€ ìš”ì²­ ë° ì‘ë‹µ ê°ì²´ ê°€ì ¸ì˜¤ê¸°
# 'https://ai-bigdataanalysis.blogspot.com/2024/03/html.html' í˜ì´ì§€ì— ì ‘ê·¼í•˜ì—¬ ì‘ë‹µ ê°ì²´ë¥¼ response ë³€ìˆ˜ì— ì €ì¥
response = urlopen('https://ai-bigdataanalysis.blogspot.com/2024/03/html.html')

# ì‘ë‹µëœ HTML ë¬¸ì„œë¥¼ ì½ì–´ì™€ì„œ ë°”ì´ë„ˆë¦¬ ë°ì´í„°ë¥¼ ë¬¸ìì—´ë¡œ ë””ì½”ë”©
html_doc = response.read().decode()

# BeautifulSoup ê°ì²´ ìƒì„± ë° HTML íŒŒì‹±
# HTML ì†ŒìŠ¤ë¥¼ BeautifulSoupì„ ì‚¬ìš©í•˜ì—¬ íŒŒì‹±í•˜ê³  soup ë³€ìˆ˜ì— ì €ì¥
soup = bs(html_doc, 'html.parser')

# íŠ¹ì • div íƒœê·¸ë¥¼ ì„ íƒí•˜ì—¬ í•´ë‹¹ ì½˜í…ì¸  ê°€ì ¸ì˜¤ê¸°
# í´ë˜ìŠ¤ê°€ 'post-body entry-content'ì¸ div íƒœê·¸ë¥¼ ì°¾ìŒ
content = soup.find("div", {"class": "post-body entry-content"})

# í…ìŠ¤íŠ¸ë¥¼ ì €ì¥í•  ë³€ìˆ˜ ì´ˆê¸°í™”
text = ""

# ì°¾ì€ div íƒœê·¸ ë‚´ë¶€ì—ì„œ ëª¨ë“  <span> íƒœê·¸ë¥¼ ìˆœíšŒí•˜ë©´ì„œ í…ìŠ¤íŠ¸ë¥¼ ì¶”ì¶œ
if content:  # contentê°€ Noneì´ ì•„ë‹ ê²½ìš°ì—ë§Œ ì‹¤í–‰
    for para in content.find_all("span"):
        # ê° <span> íƒœê·¸ì˜ í…ìŠ¤íŠ¸ë¥¼ ê°€ì ¸ì™€ì„œ ì–‘ìª½ ê³µë°± ì œê±° í›„ text ë³€ìˆ˜ì— ì¶”ê°€
        text += para.text.strip()

# ìµœì¢…ì ìœ¼ë¡œ ìˆ˜ì§‘ëœ í…ìŠ¤íŠ¸ ì¶œë ¥
print(text)
````
