### A. ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ ì „ì²˜ë¦¬
1. **ì¤‘ì•™ê°’** ëŒ€ì²´
2. **í‰ê· ê°’** ëŒ€ì²´(ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ í‰ê· ê°’ ëŒ€ì²´ í›„ **intë¡œ ë³€í™˜**í•´ì¤˜ì•¼ í•¨)

````python
# 1. ì¤‘ì•™ê°’ ëŒ€ì²´
x_train_nan[float_columns[0]].fillna(x_train_nan[float_columns[0]].median(), inplace=False).astype('int')
````

````python
# 2. í‰ê· ê°’ ëŒ€ì²´ (ì •ìˆ˜í˜•ì´ê¸° ë•Œë¬¸ì— intë¡œ ë³€í™˜)
x_train_nan[float_columns[0]].fillna(x_train_nan[float_columns[0]].mean(), inplace=False).astype('int')
````

````python
# 3. KNN Imputer
from sklearn.impute import KNNImputer

df = x_train_nan.copy()
imputer = KNNImputer(n_neighbors=5) # 3~5 ì¤‘ ì„ íƒ

df[float_columns] = imputer.fit_transform(df[float_columns])
df[float_columns] = df[float_columns].astype('int32')
````

### B. ë²”ì£¼í˜• ì»¬ëŸ¼ ì „ì²˜ë¦¬
1. ê²°ì¸¡ì¹˜ ì‚­ì œ -> ìœ íš¨í•œ ë°ì´í„°ê°€ ì‚­ì œë  ìœ„í—˜ìˆìŒ
2. **ê·¸ë£¹ë³„ ìµœë¹ˆê°’** ëŒ€ì²´
3. ìœ ì‚¬í•œ ë ˆì½”ë“œì˜ ê°’ìœ¼ë¡œ ëŒ€ì²´(**KNN Imputer**) -> ì‹¤ì œ ë°ì´í„°ì— ì „ì²˜ë¦¬ ì ìš©


````python
# 1. ê²°ì¸¡ì¹˜ ì‚­ì œ -> í˜„ì¬ëŠ” ì „ì²´ ì»¬ëŸ¼ì— ê²°ì¸¡ì¹˜ê°€ ì¡´ì¬í•˜ë„ë¡ ê²°ì¸¡ì¹˜ë¥¼ ë§ì´ ìƒì„±í•œ ìƒí™©ì´ë¯€ë¡œ ì „ì²´ ë°ì´í„°ê°€ ì‚­ì œë¨.
# ì§€ê¸ˆê³¼ ê°™ì´ ê²°ì¸¡ì¹˜ê°€ ë§ì€ ìƒí™©ì—ì„œëŠ” ë¶€ì ì ˆí•œ ë°©ë²•ì´ë¼ê³  ë³¼ ìˆ˜ ìˆìŒ
x_train_nan.dropna()
````

````python
# 2. ê·¸ë£¹ë³„ ìµœë¹ˆê°’ ëŒ€ì²´ - groupby
x_train_nan.groupby('age')['school'].transform(lambda x: x.fillna(x.mode()[0]))
````

````python
# 3. KNN Imputer
# ê²°ì¸¡ê°’ì´ ìˆëŠ” ë¬¸ìì—´ ì»¬ëŸ¼ì„ ê°€ì¥ ê°€ê¹Œìš´ ì´ì›ƒì˜ ê°’ìœ¼ë¡œ ëŒ€ì²´
# ë¬¸ìì—´ì€ ìˆ˜ì¹˜ë¡œ ë³€í™˜í•œ ë’¤ ì ìš©

# df = x_train_nan.copy()

from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# ìˆ«ì ë³€í™˜
label_encoders = {}  # ê° ì»¬ëŸ¼ë³„ LabelEncoder ì €ì¥
for col in object_columns: # ë¬¸ìì—´ ì»¬ëŸ¼
    le = LabelEncoder()
    # ê²°ì¸¡ì¹˜ ì œì™¸í•œ ë°ì´í„°ë§Œ LabelEncoderì— ì ìš© -> ê²°ì¸¡ì¹˜ë¥¼ ì±„ìš°ê¸° ìœ„í•´, ê²°ì¸¡ì¹˜ëŠ” ë¼ë²¨ë¡œ ì¸ì‹í•˜ë©´ ì•ˆë¨.
    non_nan_mask = df[col].notnull()  # ê²°ì¸¡ì¹˜ê°€ ì•„ë‹Œ í–‰ í•„í„°
    df.loc[non_nan_mask, col + '_encoded'] = le.fit_transform(df.loc[non_nan_mask, col])

    label_encoders[col] = le  # ê° ì»¬ëŸ¼ì˜ LabelEncoder ì €ì¥ -> ê° ì»¬ëŸ¼ë§ˆë‹¤ labelì´ ë‹¤ë¥´ê¸° ë•Œë¬¸ì— ê°ê° ì €ì¥í•´ì•¼ í•¨
````

````python
# ì—­ë³€í™˜ì„ í†µí•´ ì›ë˜ ë¬¸ìì—´ë¡œ ë³µì›(í•„ìš”ì‹œ)
for col in object_columns:
    # print(col)
    df[col] = label_encoders[col].inverse_transform(df[col + '_encoded'].astype('int'))
````

## 2-3. ë°ì´í„° ì´ìƒì¹˜ ëŒ€ì²´(Interpolate)
- IQRì„ ì‚¬ìš©í•´ ì´ìƒì¹˜ í™•ì¸
- interpolateë¥¼ ì‚¬ìš©í•´ ì„ í˜• ë³´ê°„ ì‚¬ìš©

````python
# IQR(ì‚¬ë¶„ìœ„ìˆ˜ ë²”ìœ„) ë°©ë²•ì„ ì‚¬ìš©í•˜ì—¬ ì´ìƒì¹˜ë¥¼ ì‹ë³„
Q1 = data['SUNACTIVITY'].quantile(0.25)  # ì²« ë²ˆì§¸ ì‚¬ë¶„ìœ„ìˆ˜(25ë²ˆì§¸ ë°±ë¶„ìœ„ìˆ˜)
Q3 = data['SUNACTIVITY'].quantile(0.75)  # ì„¸ ë²ˆì§¸ ì‚¬ë¶„ìœ„ìˆ˜(75ë²ˆì§¸ ë°±ë¶„ìœ„ìˆ˜)
IQR = Q3 - Q1  # ì‚¬ë¶„ìœ„ìˆ˜ ë²”ìœ„(IQR)

# ì´ìƒì¹˜ ê¸°ì¤€ ì •ì˜
lower_bound = Q1 - 1.5 * IQR  # í•˜í•œ
upper_bound = Q3 + 1.5 * IQR  # ìƒí•œ

# ì´ìƒì¹˜ íƒì§€
outliers = (data['SUNACTIVITY'] < lower_bound) | (data['SUNACTIVITY'] > upper_bound)  # ì´ìƒì¹˜ ì¡°ê±´ ì„¤ì •

# ì´ìƒì¹˜ë¥¼ NaNìœ¼ë¡œ ëŒ€ì²´í•˜ì—¬ ë³´ê°„(interpolation) ì¤€ë¹„
data_interpolated = data.copy()  # ì›ë³¸ ë°ì´í„° ë³µì‚¬
data_interpolated.loc[outliers, 'SUNACTIVITY'] = None  # ì´ìƒì¹˜ë¥¼ NaNìœ¼ë¡œ ëŒ€ì²´

# ê²°ì¸¡ì¹˜ ë³´ê°„ (ì„ í˜• ë³´ê°„ ì‚¬ìš©)
data_interpolated['SUNACTIVITY'] = data_interpolated['SUNACTIVITY'].interpolate()

# ë¹„êµë¥¼ ìœ„í•œ ì‹œê°í™”
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['SUNACTIVITY'], label='Original Data', alpha=0.6)
plt.plot(data_interpolated.index, data_interpolated['SUNACTIVITY'], label='Interpolated Data', linestyle='--', color='orange')
plt.legend()
plt.title("Sunspot Activity Before and After Outlier Interpolation", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Sunspot Activity", fontsize=12)
plt.show()
````

````python
# ì´ìƒì¹˜ ì œê±° í›„, boxplot í™•ì¸
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_interpolated['SUNACTIVITY'], orient='h', color='skyblue')
plt.title('Boxplot of Interpolated Sunspot Activity', fontsize=14)
plt.xlabel('Sunspot Activity', fontsize=12)
plt.show()
````

## 2-4. SARIMAë¥¼ ì‚¬ìš©í•´ ëª¨ë¸ë§
- ì´ìƒì¹˜ ì „ì²˜ë¦¬ ì „ í›„ ì„±ëŠ¥ í™•ì¸
- íŒŒë¼ë¯¸í„° ì„¤ëª…
   - order=(1, 1, 1):

      - ê³¼ê±° ê°’ 1ë‹¨ê³„ê¹Œì§€(AR1)ë¥¼ í¬í•¨.<BR>
      - ë°ì´í„°ê°€ 1ì°¨ ì°¨ë¶„ìœ¼ë¡œ ì•ˆì •í™”ë¨(d=1).<bR>
      - ì”ì°¨ì˜ ê³¼ê±° 1ë‹¨ê³„ê¹Œì§€(MA1)ë¥¼ í¬í•¨.<BR>

  - seasonal_order=(1, 1, 1, 12):

    - ê³„ì ˆì  ì£¼ê¸° m=12(ì˜ˆ: 1ë…„ ë‹¨ìœ„ ì£¼ê¸°).<BR>
    - ê³„ì ˆì  ìê¸°íšŒê·€, ì°¨ë¶„, ì´ë™ í‰ê·  ê° 1ë‹¨ê³„ê¹Œì§€ í¬í•¨.<BR>
    
[ëª¨ë¸ ì„±ëŠ¥]
- RMSE(Root Mean Squared Error)
  - MSEì˜ ì œê³±ê·¼ìœ¼ë¡œ, MSEë³´ë‹¤ ë‹¨ìœ„ê°€ ì‹¤ì œ ë°ì´í„°ì˜ ë‹¨ìœ„ì™€ ë™ì¼
  - ê°’ì´ ì‘ì„ìˆ˜ë¡ ëª¨ë¸ ì„±ëŠ¥ì´ ì¢‹ìŒ

````python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
````

````python
# SARIMAë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµ ë° ì˜ˆì¸¡í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def train_and_predict_sarima(train_data, test_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)):
    # SARIMA ëª¨ë¸ í•™ìŠµ
    model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)
    fitted_model = model.fit(disp=False)  # ëª¨ë¸ í”¼íŒ…

    # í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡
    predictions = fitted_model.forecast(steps=len(test_data))

    # RMSE ê³„ì‚°
    rmse = np.sqrt(mean_squared_error(test_data, predictions))
    return predictions, rmse

# ë°ì´í„°ì…‹ì„ í•™ìŠµ(train)ê³¼ í…ŒìŠ¤íŠ¸(test)ë¡œ ë¶„ë¦¬
train = data.iloc[:-5]['SUNACTIVITY']  # í•™ìŠµ ë°ì´í„°
test = data.iloc[-5:]['SUNACTIVITY']  # í…ŒìŠ¤íŠ¸ ë°ì´í„°

# ì›ë³¸ ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµ ë° ì˜ˆì¸¡
pred_original, rmse_original = train_and_predict_sarima(train, test)

# ë³´ê°„ëœ ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµ ë° ì˜ˆì¸¡
train_interpolated = data_interpolated.iloc[:-5]['SUNACTIVITY']  # ë³´ê°„ëœ í•™ìŠµ ë°ì´í„°
test_interpolated = data_interpolated.iloc[-5:]['SUNACTIVITY']  # ë³´ê°„ëœ í…ŒìŠ¤íŠ¸ ë°ì´í„°
pred_interpolated, rmse_interpolated = train_and_predict_sarima(train_interpolated, test_interpolated)
````

- ì‹œê°í™” ì§„í–‰

````python
# ì‹¤ì œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì™€ ì˜ˆì¸¡ ë°ì´í„°ë¥¼ ë¹„êµí•˜ëŠ” ì‹œê°í™”
plt.figure(figsize=(12, 8))

# ì‹¤ì œ í…ŒìŠ¤íŠ¸ ë°ì´í„°
plt.plot(test.index, test, label='Actual Data', color='black', linewidth=2)

# ì›ë³¸ ë°ì´í„°ì™€ ë³´ê°„ëœ ë°ì´í„°ì˜ ì˜ˆì¸¡ê°’
plt.plot(test.index, pred_original, label='Predictions (Original Data)', linestyle='--', color='red')
plt.plot(test.index, pred_interpolated, label='Predictions (Interpolated Data)', linestyle='-.', color='blue')

# plot ì„¤ì •
plt.legend()
plt.title("SARIMA Predictions: Original vs Interpolated Data", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Sunspot Activity", fontsize=12)
plt.grid(alpha=0.5)
plt.show()

# RMSE values í”„ë¦°íŠ¸
print(f"RMSE (Original Data): {rmse_original}")
print(f"RMSE (Interpolated Data): {rmse_interpolated}")
````

- ì´ ë¼ì´ë¸ŒëŸ¬ë¦¬ëŠ” ë¬¸ë§¥ì  ì˜ë¯¸ë¥¼ í•™ìŠµí•˜ê¸° ë•Œë¬¸ì— ì¤‘ì˜ì  ë¬¸ì œ í•´ê²°í•¨

````python
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch

# ëª¨ë¸ ë° í† í¬ë‚˜ì´ì € ë¡œë“œ
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForMaskedLM.from_pretrained("bert-base-uncased")

# ë¬¸ì¥ ì •ì˜
sentences = [
    "I went to the bank to deposit money.", # bank: ì€í–‰
    "She sat by the bank of the river." # bank: ê°•ë‘‘(ì œë°©)
]

# Masked Token ì¤€ë¹„ (bankë¥¼ [MASK]ë¡œ ëŒ€ì²´)
masked_sentences = [
    "I went to the [MASK] to deposit money.",
    "She sat by the [MASK] of the river."
]
````

````python
# ë¬¸ì¥ì„ í† í°í™”í•˜ê³  ëª¨ë¸ ì…ë ¥ ì¤€ë¹„
inputs = tokenizer(masked_sentences, return_tensors="pt", padding=True, truncation=True)
with torch.no_grad():
    outputs = model(**inputs)

# [MASK] í† í°ì˜ ìœ„ì¹˜ ì¸ë±ìŠ¤
mask_token_index = (inputs.input_ids == tokenizer.mask_token_id).nonzero(as_tuple=True)

# ëª¨ë¸ì˜ ì¶œë ¥ logitsì—ì„œ [MASK] ìœ„ì¹˜ì˜ í† í°ë§Œ ì¶”ì¶œ
mask_token_logits = outputs.logits[mask_token_index]

# ê°€ì¥ ë†’ì€ í™•ë¥ ì˜ ë‹¨ì–´ë¥¼ ì˜ˆì¸¡
top_tokens = torch.topk(mask_token_logits, k=5, dim=1).indices

# ê²°ê³¼ ì¶œë ¥
decoded_predictions = [[tokenizer.decode([token]) for token in tokens] for tokens in top_tokens]

# ê²°ê³¼ ì •ë¦¬
for i, sentence in enumerate(masked_sentences):
    print(f"Original Sentence: {sentences[i]}")
    print(f"Masked Sentence: {sentence}")
    print(f"Top Predictions for '[MASK]': {decoded_predictions[i]}")
    print()
````
