## 1-3. 결측치 임의로 만들기

- **결측값이 존재하지 않아, 임의로 결측값을 만들어줌.**
- 결측값이 존재하는 데이터: **x_train_nan** 로 지정

````python
num_missing = 5000  # 결측값 갯수 정하기

# 랜덤 시드 고정
np.random.seed(42)

# 결측값 랜덤으로 주기
col = x_train.columns.drop('StudentID')
missing_indices = np.random.choice(x_train.index, size=num_missing, replace=True)
missing_features = np.random.choice(col, size=num_missing, replace=True)

# 결측값이 존재하는 데이터로 만들어주기
x_train_nan = x_train.copy()

for idx, feature in zip(missing_indices, missing_features):
    x_train_nan.loc[idx, feature] = np.nan

# 결측값 갯수 확인
x_train_missing_summary = x_train_nan.isnull().sum()
x_train_missing_summary
````

- 결측치 시각화

````python
plt.figure(figsize=(20,3))
sns.barplot(x_train_missing_summary, palette='husl')
plt.xticks(rotation=45)
plt.title("Random Missing Data")
````

## 1-4. 데이터 결측치 전처리
- 컬럼 타입이 실수(정수)나 문자열일 때, 전처리 방법이 다름
- 전처리 할 데이터 -> **df** = x_train_nan.copy()

````python
# 실수형(float) 컬럼 추출 -> 원래 정수였으나, 결측치를 주면서 실수형으로 바뀜
float_columns = x_train_nan.select_dtypes(include=['float64']).columns

# 문자열(object) 컬럼 추출
object_columns = x_train_nan.select_dtypes(include=['object']).columns

# 출력
print("실수형 컬럼:")
print(float_columns)

print("\n문자열 컬럼:")
print(object_columns)
````

### A. 수치형 컬럼 전처리
1. **중앙값** 대체
2. **평균값** 대체(수치형 컬럼 평균값 대체 후 **int로 변환**해줘야 함)
3. **KNN Imputer** -> 실제 데이터에 전처리 적용

````python
# 1. 중앙값 대체
x_train_nan[float_columns[0]].fillna(x_train_nan[float_columns[0]].median(), inplace=False).astype('int')
````

````python
# 2. 평균값 대체 (정수형이기 때문에 int로 변환)
x_train_nan[float_columns[0]].fillna(x_train_nan[float_columns[0]].mean(), inplace=False).astype('int')
````

````python
# 3. KNN Imputer
from sklearn.impute import KNNImputer

df = x_train_nan.copy()
imputer = KNNImputer(n_neighbors=5) # 3~5 중 선택

df[float_columns] = imputer.fit_transform(df[float_columns])
df[float_columns] = df[float_columns].astype('int32')
````

````python
df[float_columns] = df[float_columns].astype('int32')
````

````python
# 결측치가 채워졌는지 확인
df[float_columns].info()
````

### B. 범주형 컬럼 전처리
1. 결측치 삭제 -> 유효한 데이터가 삭제될 위험있음
2. **그룹별 최빈값** 대체
3. 유사한 레코드의 값으로 대체(**KNN Imputer**) -> 실제 데이터에 전처리 적용


````python
# 1. 결측치 삭제 -> 현재는 전체 컬럼에 결측치가 존재하도록 결측치를 많이 생성한 상황이므로 전체 데이터가 삭제됨.
# 지금과 같이 결측치가 많은 상황에서는 부적절한 방법이라고 볼 수 있음
x_train_nan.dropna()
````

````python
# 2. 그룹별 최빈값 대체 - groupby
x_train_nan.groupby('age')['school'].transform(lambda x: x.fillna(x.mode()[0]))
````

````python
# 3. KNN Imputer
# 결측값이 있는 문자열 컬럼을 가장 가까운 이웃의 값으로 대체
# 문자열은 수치로 변환한 뒤 적용

# df = x_train_nan.copy()

from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# 숫자 변환
label_encoders = {}  # 각 컬럼별 LabelEncoder 저장
for col in object_columns: # 문자열 컬럼
    le = LabelEncoder()
    # 결측치 제외한 데이터만 LabelEncoder에 적용 -> 결측치를 채우기 위해, 결측치는 라벨로 인식하면 안됨.
    non_nan_mask = df[col].notnull()  # 결측치가 아닌 행 필터
    df.loc[non_nan_mask, col + '_encoded'] = le.fit_transform(df.loc[non_nan_mask, col])

    label_encoders[col] = le  # 각 컬럼의 LabelEncoder 저장 -> 각 컬럼마다 label이 다르기 때문에 각각 저장해야 함
````

````python
knn_imputer = KNNImputer(n_neighbors=3)

# 결측값 대체
encoded_columns = [col for col in df.columns if '_encoded' in col]
df[encoded_columns] = knn_imputer.fit_transform(df[encoded_columns])
````

````python
df[encoded_columns].info() # 결측값 채우기 완료
````

````python
# 역변환을 통해 원래 문자열로 복원(필요시)
for col in object_columns:
    # print(col)
    df[col] = label_encoders[col].inverse_transform(df[col + '_encoded'].astype('int'))
````

````python
df_c = df.copy()

df_c.drop(encoded_columns, axis=1, inplace=True) # encoded 삭제해줌
````

````python
df_c.info() # 결측치 처리 완료
````

## 1-5. Random Forest 회귀 성능 비교
- MSE(Mean Squared Error)
  -  예측값과 실제값의 차이를 제곱한 후 평균을 계산한 값
  - 값이 작을수록 모델 성능이 좋은
- RMSE(Root Mean Squared Error)
  - MSE의 제곱근으로, MSE보다 단위가 실제 데이터의 단위와 동일
  - 값이 작을수록 모델 성능이 좋음
- MAE(Mean Absolute Error)
  -  예측값과 실제값의 차이를 절대값을 평균한 값
  - 값이 작을수록 모델 성능이 좋음
- R^2
  - 모델이 데이터를 얼마나 잘 설명하는지 나타내는 지표
  - 0~1사이의 값으로 1에 가까울수록 성능이 좋음

### RandomForest 회귀 성능 확인(결측치 있을때)
- 결측치 있는 데이터: x_train_nan

````python
# 학생 아이디는 학습에 필요 없는 컬럼이기 때문에 drop 시킴.
drop_col = ['StudentID']

x_train_drop = x_train_nan.drop(columns = drop_col) # 학습 데이터
x_test_drop = x_test.drop(columns = drop_col) # 테스트 데이터
y_train_target = y_train['G3'] # 타겟값
y_test_target = y_test['G3']

# 범주형 데이터를 수치형 데이터로 변환
x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)[x_train_dum.columns]

# randomforest 회귀 사용
rf_nan = RandomForestRegressor(random_state = 42)
rf_nan.fit(x_train_dum,y_train_target)


# 모델평가
# mse : mean_squared_error / mae : mean_absolute_error / r2 score : r2_score
# rmse : root_mean_squerd_error -> 패키지가 없음 np.sqrt(mean_squared_error) 해줘야함

# 최종 모델 예측
y_test_pred = rf_nan.predict(x_test_dum)
#mse
print('test mse' ,mean_squared_error(y_test_target,y_test_pred).round(3))

#mae
print('test mae' ,mean_absolute_error(y_test_target,y_test_pred).round(3))

#rmse
print('test rmse' ,np.sqrt(mean_squared_error(y_test_target,y_test_pred)).round(3))

#r2
print('test r2 score' ,r2_score(y_test_target,y_test_pred))
````

### RandomForest 회귀 성능 확인(결측치 전처리 후)

- 결측치 처리한 데이터: df_c
- 결측치 채운 성능

````python
# 학생 아이디는 필요없는 컬럼이기 때문에 drop 시킴.
drop_col = ['StudentID']

x_train_drop = df_c.drop(columns = drop_col) # 학습 데이터
x_test_drop = x_test.drop(columns = drop_col) # 테스트 데이터
y_train_target = y_train['G3'] # 타겟값
y_test_target = y_test['G3']

# 범주형 데이터를 수치형 데이터로 변환
x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)[x_train_dum.columns]

# randomforest 회귀 사용
rf_last = RandomForestRegressor(random_state = 42)
rf_last.fit(x_train_dum,y_train_target)

# 모델평가
# mse : mean_squared_error / mae : mean_absolute_error  / mape : mean_absolute_percentage_error
# rmse : root_mean_squerd_error -> 패키지 없음 np.sqrt(mean_squared_error) 해줘야함

# 최종 모델 예측
y_test_pred = rf_last.predict(x_test_dum)
#mse
print('test mse' ,mean_squared_error(y_test_target,y_test_pred).round(3))

#mae
print('test mae' ,mean_absolute_error(y_test_target,y_test_pred).round(3))

#rmse
print('test rmse' ,np.sqrt(mean_squared_error(y_test_target,y_test_pred)).round(3))

#r2
print('test r2 score' ,r2_score(y_test_target,y_test_pred))
````

- 랜덤포레스트(Random Forest)는 여러 개의 결정 트리를 조합하여 예측하는 모델로, 결측치로 인한 영향을 상대적으로 잘 완화할 수 있는 특성을 가지고 있습니다. 샘플링(Bootstrap Sampling)을 통해 일부 결측치가 있더라도 데이터를 활용할 수 있고, 앙상블 기법으로 결측치의 영향을 분산시키기 때문입니다.

## 2-3. 데이터 이상치 대체(Interpolate)
- IQR을 사용해 이상치 확인
- interpolate를 사용해 선형 보간 사용

````python
# IQR(사분위수 범위) 방법을 사용하여 이상치를 식별
Q1 = data['SUNACTIVITY'].quantile(0.25)  # 첫 번째 사분위수(25번째 백분위수)
Q3 = data['SUNACTIVITY'].quantile(0.75)  # 세 번째 사분위수(75번째 백분위수)
IQR = Q3 - Q1  # 사분위수 범위(IQR)

# 이상치 기준 정의
lower_bound = Q1 - 1.5 * IQR  # 하한
upper_bound = Q3 + 1.5 * IQR  # 상한

# 이상치 탐지
outliers = (data['SUNACTIVITY'] < lower_bound) | (data['SUNACTIVITY'] > upper_bound)  # 이상치 조건 설정

# 이상치를 NaN으로 대체하여 보간(interpolation) 준비
data_interpolated = data.copy()  # 원본 데이터 복사
data_interpolated.loc[outliers, 'SUNACTIVITY'] = None  # 이상치를 NaN으로 대체

# 결측치 보간 (선형 보간 사용)
data_interpolated['SUNACTIVITY'] = data_interpolated['SUNACTIVITY'].interpolate()

# 비교를 위한 시각화
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['SUNACTIVITY'], label='Original Data', alpha=0.6)
plt.plot(data_interpolated.index, data_interpolated['SUNACTIVITY'], label='Interpolated Data', linestyle='--', color='orange')
plt.legend()
plt.title("Sunspot Activity Before and After Outlier Interpolation", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Sunspot Activity", fontsize=12)
plt.show()
````

````python
# 이상치 제거 후, boxplot 확인
plt.figure(figsize=(8, 6))
sns.boxplot(data=data_interpolated['SUNACTIVITY'], orient='h', color='skyblue')
plt.title('Boxplot of Interpolated Sunspot Activity', fontsize=14)
plt.xlabel('Sunspot Activity', fontsize=12)
plt.show()
````

## 2-4. SARIMA를 사용해 모델링
- 이상치 전처리 전 후 성능 확인
- 파라미터 설명
   - order=(1, 1, 1):

      - 과거 값 1단계까지(AR1)를 포함.<BR>
      - 데이터가 1차 차분으로 안정화됨(d=1).<bR>
      - 잔차의 과거 1단계까지(MA1)를 포함.<BR>

  - seasonal_order=(1, 1, 1, 12):

    - 계절적 주기 m=12(예: 1년 단위 주기).<BR>
    - 계절적 자기회귀, 차분, 이동 평균 각 1단계까지 포함.<BR>
    
[모델 성능]
- RMSE(Root Mean Squared Error)
  - MSE의 제곱근으로, MSE보다 단위가 실제 데이터의 단위와 동일
  - 값이 작을수록 모델 성능이 좋음

````python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
````

````python
# SARIMA를 사용하여 학습 및 예측하는 함수 정의
def train_and_predict_sarima(train_data, test_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)):
    # SARIMA 모델 학습
    model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order)
    fitted_model = model.fit(disp=False)  # 모델 피팅

    # 테스트 데이터에 대한 예측
    predictions = fitted_model.forecast(steps=len(test_data))

    # RMSE 계산
    rmse = np.sqrt(mean_squared_error(test_data, predictions))
    return predictions, rmse

# 데이터셋을 학습(train)과 테스트(test)로 분리
train = data.iloc[:-5]['SUNACTIVITY']  # 학습 데이터
test = data.iloc[-5:]['SUNACTIVITY']  # 테스트 데이터

# 원본 데이터를 사용하여 학습 및 예측
pred_original, rmse_original = train_and_predict_sarima(train, test)

# 보간된 데이터를 사용하여 학습 및 예측
train_interpolated = data_interpolated.iloc[:-5]['SUNACTIVITY']  # 보간된 학습 데이터
test_interpolated = data_interpolated.iloc[-5:]['SUNACTIVITY']  # 보간된 테스트 데이터
pred_interpolated, rmse_interpolated = train_and_predict_sarima(train_interpolated, test_interpolated)
````

- 시각화 진행

````python
# 실제 테스트 데이터와 예측 데이터를 비교하는 시각화
plt.figure(figsize=(12, 8))

# 실제 테스트 데이터
plt.plot(test.index, test, label='Actual Data', color='black', linewidth=2)

# 원본 데이터와 보간된 데이터의 예측값
plt.plot(test.index, pred_original, label='Predictions (Original Data)', linestyle='--', color='red')
plt.plot(test.index, pred_interpolated, label='Predictions (Interpolated Data)', linestyle='-.', color='blue')

# plot 설정
plt.legend()
plt.title("SARIMA Predictions: Original vs Interpolated Data", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Sunspot Activity", fontsize=12)
plt.grid(alpha=0.5)
plt.show()

# RMSE values 프린트
print(f"RMSE (Original Data): {rmse_original}")
print(f"RMSE (Interpolated Data): {rmse_interpolated}")
````

````python
# 소문자 변환
def to_lowercase(text):
    return text.lower()

# 오탈자
def correct_spelling_textblob(text):
    blob = TextBlob(text)
    return str(blob.correct())

# 중복 제거 함수
def remove_redundancy(text):
    return re.sub(r'\b(\w+)\b(?:.*?\b\1\b)+', r'\1', text)

# 특수문자 제거
def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)
````

# 추가 학습) 토큰화 및 문맥적 의미
- Hugging Face의 transformers 라이브러리
- 이 라이브러리는 문맥적 의미를 학습하기 때문에 중의적 문제 해결함

````python
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch

# 모델 및 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForMaskedLM.from_pretrained("bert-base-uncased")

# 문장 정의
sentences = [
    "I went to the bank to deposit money.", # bank: 은행
    "She sat by the bank of the river." # bank: 강둑(제방)
]

# Masked Token 준비 (bank를 [MASK]로 대체)
masked_sentences = [
    "I went to the [MASK] to deposit money.",
    "She sat by the [MASK] of the river."
]
````

````python
# 문장을 토큰화하고 모델 입력 준비
inputs = tokenizer(masked_sentences, return_tensors="pt", padding=True, truncation=True)
with torch.no_grad():
    outputs = model(**inputs)

# [MASK] 토큰의 위치 인덱스
mask_token_index = (inputs.input_ids == tokenizer.mask_token_id).nonzero(as_tuple=True)

# 모델의 출력 logits에서 [MASK] 위치의 토큰만 추출
mask_token_logits = outputs.logits[mask_token_index]

# 가장 높은 확률의 단어를 예측
top_tokens = torch.topk(mask_token_logits, k=5, dim=1).indices

# 결과 출력
decoded_predictions = [[tokenizer.decode([token]) for token in tokens] for tokens in top_tokens]

# 결과 정리
for i, sentence in enumerate(masked_sentences):
    print(f"Original Sentence: {sentences[i]}")
    print(f"Masked Sentence: {sentence}")
    print(f"Top Predictions for '[MASK]': {decoded_predictions[i]}")
    print()
````
